```python
import pandas as pd
```

```python
salaries = pd.read_csv('salary_data.csv')
```

```python
salaries
```

```python
salaries.plot.scatter(x='YearsExperience', y='Salary')
```

```python
from sklearn import linear_model

X = salaries[['YearsExperience']].to_numpy()
X.shape
```

```python
Y = salaries[['Salary']].to_numpy()
Y.shape
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.predict(5) # nera apmokytas dar
```

```python
model.fit(X, Y)
```

```python
model.predict( [[5], [10], [1.1], [1.2] ] )
```

```python
salaries
```

```python
predictions = model.predict(X)
predictions
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y)
plt.plot(X, predictions, c='r')
```
