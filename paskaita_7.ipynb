```python
def my_function(row):
    return str(row['country']) + ' ' + str(row['province'])
```

## drop NaNs

```python
wine_df_cleaned = wine_df.drop_duplicates()
wine_df_cleaned
```

```python
wine_df_cleaned[ wine_df_cleaned['price'].isnull() ].shape
```

```python
wine_df_cleaned = wine_df_cleaned.dropna(how='any', subset=['price', 'province']) # drop rows with NaN in price, province columns
```

## groupby

```python
grouped = wine_df_cleaned.groupby(['country', 'province'])[['points', 'price']].mean()
grouped
```
